## Gruplama Islemleri

```python
df = pd.read_csv("train.csv")
```

```python
df[df.Sex == "female"].Age.mean()
```

```python
df.groupby("Sex")["Age"].mean()
```

```python
df.groupby("Sex").agg({"Age":"mean", "Survived":"mean"})
```

```python
df.groupby(["Sex", "Pclass"]).agg({"Age":"mean", "Survived":"mean"})
```

```python
df["Sex"].value_counts()
```

```python
df.groupby(["Sex", "Pclass", "Embarked"]).agg({"Age":"mean", "Survived":"mean"})
```

```python
df.groupby("Sex").agg({"Age":[min, np.median, max], "Survived":"mean"})
```
